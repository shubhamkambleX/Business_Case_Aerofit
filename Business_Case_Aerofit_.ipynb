```python
# Computing Marginal & Condinal Probability
```

```python
pd.concat(
    [
        df['Product'].value_counts(),
        df['Product'].value_counts(normalize=True)
    ],
    keys=['counts','Marginal_prob'],
    axis=1
)
```

```python
# Probability of each given gender
```

```python
def p_prod_given_gender(gender,print_marginal=False):
  if gender != "Female" and gender != "Male":
    return "Invalid gender"
  df_1 = pd.crosstab(index=df['Gender'],columns=df['Product'])
  p_781 = df_1['KP781'][gender]/df_1.loc[gender].sum()
  p_481 = df_1['KP481'][gender]/df_1.loc[gender].sum()
  p_281 = df_1['KP281'][gender]/df_1.loc[gender].sum()

  if print_marginal:
    print(f"p(Male): {round(df_1.loc['Male'].sum()/len(df),2)}")
    print(f"p(Female): {round(df_1.loc['Female'].sum()/len(df),2)}")

  print(f"(KP781/{gender}):{round(p_781,2)}")
  print(f"(KP481/{gender}):{round(p_481,2)}")
  print(f"(KP281/{gender}):{round(p_281,2)}")

p_prod_given_gender('Male',True)
p_prod_given_gender('Female')
```

```python
# Probability of each product given Maritial Status
```

```python
def p_prod_given_MaritalStatus(status,print_marginal=False):
  if status != "Single" and status != "Partnered":
    return "Invalid MaritalStatus value"
  df_1 = pd.crosstab(index=df['MaritalStatus'],columns=df['Product'])
  p_781 = df_1['KP781'][status]/df_1.loc[status].sum()
  p_481 = df_1['KP481'][status]/df_1.loc[status].sum()
  p_281 = df_1['KP281'][status]/df_1.loc[status].sum()

  if print_marginal:
    print(f"p(Single): {round(df_1.loc['Single'].sum()/len(df),2)}")
    print(f"p(Partnered): {round(df_1.loc['Partnered'].sum()/len(df),2)}")

  print(f"(KP781/{status}):{round(p_781,2)}")
  print(f"(KP481/{status}):{round(p_481,2)}")
  print(f"(KP281/{status}):{round(p_281,2)}")

p_prod_given_MaritalStatus('Single',True)
p_prod_given_MaritalStatus('Partnered')
```
